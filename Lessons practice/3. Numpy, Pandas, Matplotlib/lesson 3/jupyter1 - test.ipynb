# Magic Keywords

```python
def solution(s):
    # Your code here
    count = 0
    for i in range(len(s)):
        if s[i] == '>':
            for j in range(i,len(s)):
                if s[j] == '<':
                    count+=2
    return count
```

```python
solution("<<>><")
```

```python
%timeit solution("<<>><")
```
